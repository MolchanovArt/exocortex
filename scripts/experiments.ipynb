```python
import sys
from pathlib import Path

# Add the src directory to Python path so we can import exocortex modules
project_root = Path().resolve().parent  # Go up from scripts/ to exocortex/
src_path = project_root / "src"
if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))

from exocortex.core.config import config
from exocortex.integrations.telegram_client import _fetch_recent_messages_async

print("Configured TELEGRAM_TARGET_CHAT_ID:", config.telegram_target_chat_id)

# In Jupyter notebooks, we can await directly since there's already an event loop
msgs = await _fetch_recent_messages_async(limit=20)

print(f"Fetched {len(msgs)} messages")

for m in msgs:
    print("----")
    print("chat_id:", m.chat_id)
    print("message_id:", m.message_id)
    print("text:", repr(m.text))
    print("timestamp:", m.timestamp)
```
